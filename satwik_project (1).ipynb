#Gaussian Naive Bayes

Gaussian Naive Bayes class

```python
class GaussianNaiveBayes:
    def fit(self, x_train, y_train):
        self.uniq = y_train.unique()
        self.mean = {}
        self.var = {}
        self.prior = {}
        for uni in self.uniq:
            x_uni = x_train[y_train == uni]
            self.mean[uni] = x_uni.mean()
            self.var[uni] =  x_uni.var()
            self.prior[uni] = x_uni.shape[0] / x_train.shape[0]

    def prob(self, x, m, v):
        expo = np.exp(-((x - m) ** 2 / (2 * v)))
        return np.prod((1 / (np.sqrt(2 * np.pi * v))) * expo)

    def predict(self, x_test):
        pre = []
        for x in x_test.values:
            posteriors = []
            for uni in self.uniq:
                likelihood = self.prob(x, self.mean[uni], self.var[uni])
                posterior = likelihood * self.prior[uni]
                posteriors.append(posterior)
            pre.append(self.uniq[np.argmax(posteriors)])
        return pre

```
